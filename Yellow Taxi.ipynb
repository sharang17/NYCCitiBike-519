```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
yellow_taxi_df = pd.read_csv('yellow_tripdata_2016-06.csv')
yellow_taxi_df.info()
```

```python
yellow_taxi_df['tpep_pickup_datetime']= pd.to_datetime(yellow_taxi_df['tpep_pickup_datetime'])
yellow_taxi_df['tpep_dropoff_datetime']= pd.to_datetime(yellow_taxi_df['tpep_dropoff_datetime'])
```

```python
yellow_taxi_df.info()
```

```python
pickup_df = yellow_taxi_df
pickup_df['ride_count'] = np.ones(len(yellow_taxi_df.index))
pickup_df.index=pd.to_datetime(pickup_df['tpep_pickup_datetime'],unit='s')
pickup_df.resample('h', how='sum', axis=0)
```

```python
pickup_hist_df = pickup_df[['tpep_pickup_datetime', 'ride_count']]
pickup_result = pickup_hist_df.groupby([pickup_hist_df["tpep_pickup_datetime"].dt.month, pickup_hist_df["tpep_pickup_datetime"].dt.hour]).sum()
pickup_result.index.names = ["Month", "Hour"]
pickup_result.to_csv('pickup_result.csv')
```

```python
dropoff_df = yellow_taxi_df
dropoff_df['ride_count'] = np.ones(len(yellow_taxi_df.index))
dropoff_df.index=pd.to_datetime(dropoff_df['tpep_dropoff_datetime'], format = '%m/%d/%Y %H:%M:%S')
dropoff_df.resample('h', axis=0).sum()
#dropoff_df.fillna(value=1)
```

```python
dropoff_hist_df = dropoff_df[['tpep_dropoff_datetime', 'ride_count']]
result = dropoff_hist_df.groupby([dropoff_hist_df["tpep_dropoff_datetime"].dt.month, dropoff_hist_df["tpep_dropoff_datetime"].dt.hour]).sum()
result.index.names = ["Month", "Hour"]
```

```python
dropoff_hist_df = dropoff_df[['tpep_dropoff_datetime', 'ride_count']]
dropoff_hist_df.to_csv('drop_off_count.csv')
```
